```python
# 1. Using Re-index function

# Below we created 
# (a) We created a list with name as 'subjects' containing 'English', 'Language'...
# (b) 7 series of students, we have data of 7 students
# (c) We created a dataframe of all of our serieses for that we have created an empty dataframe in which we have written 
#     pd.DataFrame, columns= student1, student2, student3, student4.. and defined our indexes as subjects.
# (d) Now using our dataframe df2. we are creating or initiating values of my series student1 in Student1 column, 
#     similarly all the columns are getting filled.
```

```python
subjects= ['English', 'Language', 'Maths', 'Science', 'Social']
student1=pd.Series({'English':80, 'Language':45, 'Maths':99, 'Social':56,})
student2=pd.Series({ 'Language':77, 'Maths':69, 'Social':55})
student3=pd.Series({'English':44, 'Language':48, 'Maths':68,'Social':65})
student4=pd.Series({'English':67, 'Language':51, 'Social':79})
student5=pd.Series({'English':88, 'Language':63, 'Maths':83,'Science':69})
student6=pd.Series({'Language':48, 'Science':56,'Maths':80, 'Social':73})
student7=pd.Series({'English':82, 'Maths':90,'Science':88, 'Social':91})
df3= pd.DataFrame(columns=['Student1', 'Student2', 'Student3', 'Student4', 'Student5', 'Student6', 'Student7'], index=subjects)
df3.Student1=student1
df3.Student2=student2
df3.Student3=student3
df3.Student4=student4
df3.Student5=student5
df3.Student6=student6
df3.Student7=student7
df3
```

```python
# Using reindex function

# using reindex we add two subjects - 'Craft' and 'Training' 
df3.reindex(['Craft', 'English', 'Language', 'Maths', 'Science', 'Social','Training'])
```

```python
# Now since the new subject are filled with null values, we can use an attribute called fill_na
(a)  We are creating this reindex again, df2.reindex and at the end we are passing another attribute fill_value=0
```

```python
reindexdf3= df3.reindex(['Craft', 'English', 'Language', 'Maths', 'Science', 'Social','Training'], fill_value=0)
reindexdf3
```

```python
# (b) There is one more method which is forward filling, or when we want to fill the null values with the previous values from
# above row that is backward filling or bfill. 
```

```python
# Filling null values with 'ffill' method
# We have written again reindexdf3 

reindexdf4= df3.reindex(['Craft', 'English', 'Language', 'Maths', 'Science', 'Social','Training'], method='ffill')
reindexdf4
```

```python
# Filling null values with 'bfill' method
reindexdf5= df3.reindex(['Craft', 'English', 'Language', 'Maths', 'Science', 'Social','Training'], method='bfill')
reindexdf5
```

```python
## Ranking in Pandas

# When we want to provide ranks to certain scores or salary bands
```

```python
d= { 
    'Name':['Alisa', 'Bobby', 'Jodha', 'jack', 'raghu', 'Cathrine',
            'Alisa', 'Bobby', 'kumar','Alisa','Alex', 'Cathrine'], 
     'Subject':['Mathematics', 'Mathematics','Mathematics','Science','Science','Science','History',
                'History', 'History', 'Economics', 'Economics', 'Economics' ],
     'Score':[62,47,55,74,31,77,85,63,42,62,89,85]}

df8 = pd.DataFrame(d, columns=['Name','Subject','Score'])
df8
```

```python
# Now we define a new column for rank 
# Syntax- Name of the dataframe[Specify the new column name]= parameter on which you want to provide the rank

df8['score_ranked']= df8['Score'].rank(ascending=1)
df8
```

```python
# Ranking in descending order

df8['score_desc']= df['Score'].rank(ascending=False)
df8
```

```python
# RAnking of 2.5 and 7.5 means they are sharing the same value
```

```python
# Ranking by minimum value

df8['score_1']=df8['Score'].rank(ascending=0, method='min')
df8
```

```python
# Ranking by maximum value

df8['score_2']=df8['Score'].rank(ascending=0,method='max')
df8
```

```python
# Dense Rank

df8['Score_dense']=df8['Score'].rank(ascending=0,method='dense')
df8
```

```python
# Hierarchial Indexing or multiple indexing

# Convert particular column into index
```

```python
d= { 
    'Name':['Alisa', 'Bobby', 'Cathrine', 'Alisa', 'Bobby', 'Cathrine',
            'Alisa', 'Bobby', 'Cathrine','Alisa','Bobby', 'Cathrine'],
     'Exam':['Semester1','Semester1','Semester1','Semester1','Semester1','Semester1','Semester2','Semester2',
             'Semester2','Semester2','Semester2','Semester2'],
     'Subject':['Mathematics', 'Mathematics','Mathematics','Science','Science','Science','Mathematics', 'Mathematics','Mathematics',
                'Science','Science','Science'],
     'Score':[62,47,55,74,31,77,85,63,42,62,89,81]}

df9 = pd.DataFrame(d, columns=['Name','Exam','Subject','Score'])
df9
```

```python
# Convert column into index
# Here Exam is parent index and Subject is child index
# And Name and Score are the two columns

df10=df9.set_index(['Exam','Subject'])
df10
```

```python
df10.index
```

```python
# Swap the column( ie. Parent to child index) in hierarchial index

df10.swaplevel('Subject','Exam')
```

```python
# level=0 means sorting on the basis of parent column

df10.sort_index(level=0)
```
